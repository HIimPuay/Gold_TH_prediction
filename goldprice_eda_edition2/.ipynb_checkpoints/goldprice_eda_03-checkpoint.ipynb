## 1. Setup และการโหลดข้อมูลหลัก
โค้ดนี้จะนำเข้าไลบรารีที่จำเป็นทั้งหมด และโหลดไฟล์ feature_store.csv รวมถึงแปลงคอลัมน์ date ให้เป็น Time-Series Index

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# ตั้งค่าการแสดงผล
pd.set_option('display.max_columns', None)
plt.style.use('ggplot')
sns.set_style('whitegrid')

plt.rc('font', family='Tahoma')
plt.rcParams['axes.unicode_minus'] = False

# โหลด Feature Store (ไฟล์หลัก)
df = pd.read_csv('feature_store.csv')

# แปลง 'date' เป็น datetime และตั้งเป็น Index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

print("--- Data Loading Complete ---")
```

## 2. โครงสร้างข้อมูลและการตรวจสอบคุณภาพเบื้องต้น
ตรวจสอบชนิดข้อมูล จำนวนแถว/คอลัมน์ และ Missing Values

```python
print("--- ข้อมูลเบื้องต้น (Structure) ---")
df.info()

print("\n--- ตัวอย่างข้อมูล 5 แถวแรก (Head) ---")
print(df.head())

print("\n--- การตรวจสอบ Missing Values (NaN) ---")
# ตรวจสอบ Missing Values ที่มักจะเกิดขึ้นในแถวเริ่มต้นของ Lagged/Rolling Features
print(df.isnull().sum()[df.isnull().sum() > 0])
```

## 3. สถิติเชิงพรรณนา (Descriptive Statistics)
แสดงค่าสถิติสำหรับตัวแปรทั้งหมด โดยการ Transpose เพื่อความสะดวกในการอ่าน

```python
print("--- สถิติเชิงพรรณนา (Descriptive Statistics) ---")
# เลือกคอลัมน์หลักที่มีความสำคัญ
key_cols = ['gold', 'gold_next', 'fx', 'cpi', 'oil', 'set', 'btc'] + [col for col in df.columns if '_pct_change' in col]

# แสดงผลแบบ Transposed
print(df[key_cols].describe().T)
```

## 4. การกระจายตัวและ Outliers (Histograms & Box Plots)
แสดงภาพการกระจายตัวของตัวแปรหลักเพื่อทำความเข้าใจลักษณะของข้อมูล

```python
# เลือกคอลัมน์หลักสำหรับการวิเคราะห์การกระจายตัว
visual_cols = ['gold', 'gold_next', 'fx', 'btc', 'gold_pct_change', 'btc_pct_change']

# พล็อต Histograms (KDE)
plt.figure(figsize=(15, 10))
for i, col in enumerate(visual_cols):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}', fontsize=10)
    plt.xlabel(col, fontsize=8)
    plt.ylabel('Frequency', fontsize=8)
plt.tight_layout()
plt.savefig('eda_distribution_plots.png')
plt.show()

# พล็อต Box Plots สำหรับตรวจสอบ Outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(visual_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=df[col].dropna())
    plt.title(f'Box Plot of {col}', fontsize=10)
    plt.ylabel(col, fontsize=8)
plt.tight_layout()
plt.savefig('eda_boxplot_plots.png')
plt.show()
```

## 5. Correlation Matrix (Heatmap)
วิเคราะห์ความสัมพันธ์เชิงเส้นตรงระหว่างตัวแปรทั้งหมด โดยเน้นความสัมพันธ์กับ gold_next

```python
# เลือกคอลัมน์หลักและ Lag Features ที่เกี่ยวข้อง
correlation_cols = ['gold_next', 'gold', 'fx', 'cpi', 'oil', 'set', 'btc'] + \
                   [col for col in df.columns if '_lag1' in col or '_pct_change' in col]

corr_matrix = df[correlation_cols].corr()

# พล็อต Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5, cbar_kws={'shrink': 0.8})
plt.title('Correlation Heatmap (Key Features)', fontsize=14)
plt.tight_layout()
plt.savefig('eda_correlation_heatmap.png')
plt.show()

print("\n--- Correlation กับ Target (gold_next) ---")
# แสดงเฉพาะความสัมพันธ์กับตัวแปรเป้าหมาย
print(corr_matrix['gold_next'].sort_values(ascending=False).head(15))
```

## 6. Time-Series Plotting (Trend and Volatility)
แสดงภาพอนุกรมเวลาเพื่อดูแนวโน้มและการเปลี่ยนแปลงของความผันผวน

```python
ts_cols = ['gold', 'fx', 'btc', 'set']

plt.figure(figsize=(15, 12))
for i, col in enumerate(ts_cols):
    plt.subplot(4, 1, i + 1)
    df[col].plot(title=f'Time Series Plot of {col}', linewidth=1.5)
    plt.xlabel('Date')
    plt.ylabel(col)
plt.tight_layout()
plt.savefig('eda_time_series_plots.png')
plt.show()
```

## 7. Autocorrelation Function (ACF) Plot
ตรวจสอบการพึ่งพาอาศัยกันตามเวลา (Autocorrelation) ของตัวแปรสำคัญ

```python
# เลือกตัวแปรที่น่าสนใจ เช่น ราคาทองคำ และ อัตราการเปลี่ยนแปลงของ Bitcoin
acf_cols = ['gold', 'gold_pct_change', 'btc', 'btc_pct_change']

plt.figure(figsize=(15, 10))
for i, col in enumerate(acf_cols):
    plt.subplot(2, 2, i + 1)
    # พล็อต ACF โดยจำกัด lags ไว้ที่ 30 วัน
    plot_acf(df[col].dropna(), lags=30, ax=plt.gca(), title=f'ACF for {col}')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
plt.tight_layout()
plt.savefig('eda_acf_plots.png')
plt.show()
```

## 8. Stationarity Check (ADF Test)
ตรวจสอบคุณสมบัติ Stationarity ซึ่งจำเป็นสำหรับโมเดล ARIMA-type

```python
# Helper function สำหรับ ADF Test
def adf_test_results(series):
    result = adfuller(series.dropna(), autolag='AIC')
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'P-value: {result[1]:.4f}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value:.4f}')
    if result[1] <= 0.05:
        print("=> สรุป: ข้อมูลมีคุณสมบัติ Stationary (Reject Null Hypothesis)")
    else:
        print("=> สรุป: ข้อมูลไม่มีคุณสมบัติ Stationary (Fail to Reject Null Hypothesis)")

print("--- ADF Test Results (Stationarity Check) ---")

print("\n[Gold Price]")
adf_test_results(df['gold'])

print("\n[FX Rate]")
adf_test_results(df['fx'])

print("\n[Gold Percentage Change]")
# มักจะตรวจสอบตัวแปรที่ถูกหาผลต่างแล้ว (Differenced)
adf_test_results(df['gold_pct_change'])
```

## 9. การโหลดข้อมูล CPI ละเอียด (Feature Enrichment Prep)
โหลดและตรวจสอบไฟล์ CPI_clean_for_supabase.csv

```python
# โหลดข้อมูล CPI ละเอียด
cpi_df = pd.read_csv('CPI_clean_for_supabase.csv')

# แปลง 'date' เป็น datetime
cpi_df['date'] = pd.to_datetime(cpi_df['date'])

print("--- ข้อมูล CPI ละเอียด (CPI_clean_for_supabase.csv) ---")
cpi_df.info()

print("\n--- ตัวอย่างข้อมูล 5 แถวแรก ---")
print(cpi_df.head())
```

## 10. การวิเคราะห์ CPI เชิงลึกและการหา Correlation
วิเคราะห์ CPI ในหมวดหมู่ย่อย และตรวจสอบความสัมพันธ์กับราคาทองคำ

```python
selected_cpi_codes = ['00000', '10000', '51000']
cpi_subset = cpi_df[cpi_df['code'].astype(str).str.contains('|'.join(selected_cpi_codes))]

# --- 2. Pivot ตารางเพื่อให้หมวดหมู่เป็นคอลัมน์ ---
# ใช้ชื่อคอลัมน์ name_th จากข้อมูลที่กรองแล้ว
try:
    cpi_pivot = cpi_subset.pivot(index='date', columns='name_th', values='cpi_index')
    cpi_pivot.columns.name = None # ลบชื่อคอลัมน์ pivot
except ValueError as e:
    print(f"Error during pivot: {e}. Check if name_th is unique per date/code combination.")
    print("Skipping pivot and proceeding to check FFILL readiness.")
    # หาก Pivot ไม่ได้ แสดงว่ามีการซ้ำของ name_th ในวันที่เดียวกัน ซึ่งไม่ควรเกิดขึ้นในข้อมูล CPI

# --- 3. การแก้ไขปัญหา Mismatch Frequency: Resample และ Forward Fill (FFILL) ---
# เนื่องจาก CPI เป็นข้อมูลรายเดือน จึงต้องแปลงเป็นรายวันและเติมค่า
if not cpi_pivot.empty:
    # ใช้วิธี asfreq('D') เพื่อสร้างดัชนีรายวัน และ ffill() เพื่อเติมค่าล่าสุด
    cpi_daily = cpi_pivot.asfreq('D').ffill()
    
    print("--- ตัวอย่างข้อมูล CPI ที่ Pivot, Resample และ FFILL แล้ว (5 แถวแรก) ---")
    print(cpi_daily.head())
    
    # --- 4. Inner Join กับ Feature Store ---
    # ใช้ Inner Join ที่มีโอกาสสำเร็จสูงขึ้นมากหลังจาก FFILL
    merged_df = df.join(cpi_daily, how='inner')
    
    # --- 5. คำนวณ Correlation ---
    print("\n--- Correlation ของ CPI หมวดหมู่ย่อยกับ gold_next (หลัง FFILL) ---")
    
    # ดึงชื่อคอลัมน์ที่ถูก Pivot
    cpi_cols = cpi_daily.columns.tolist()
    
    for col in cpi_cols:
        if col in merged_df.columns and not merged_df[col].isnull().all():
             correlation_value = merged_df[col].corr(merged_df['gold_next'])
             print(f"CPI ({col}) Correlation: {correlation_value:.4f}")
        elif col in merged_df.columns:
             print(f"CPI ({col}) Correlation: Cannot calculate (All NaN after join)")
    
else:
    print("ไม่สามารถทำการ Pivot ข้อมูล CPI ได้ โปรดตรวจสอบรหัส (code) หรือความซ้ำซ้อนของข้อมูลในไฟล์ CPI")
    merged_df = pd.DataFrame() # กำหนดให้ merged_df เป็น DataFrame ว่างเพื่อไม่ให้โค้ดส่วนอื่นมีปัญหา
```

## 11. Time-Series Comparison Plot (Dual-Axis Plot)
โค้ดนี้จะเปรียบเทียบแนวโน้มของราคาทองคำ (gold) กับดัชนีราคาผู้บริโภครวม (อาหารและเครื่องดื่มไม่มีแอลกอฮอล์) โดยใช้แกน Y สองแกน (Dual-Axis) เพื่อจัดการกับ Scale ที่ต่างกัน

```python
# เปลี่ยนจาก 'รวมทุกรายการ' ไปใช้คอลัมน์ที่ยืนยันแล้วว่ามีอยู่จริง
cpi_trend_col = 'หมวดอาหารและเครื่องดื่มไม่มีแอลกอฮอล์' 
gold_col = 'gold'

if cpi_trend_col in merged_df.columns and gold_col in merged_df.columns:
    
    fig, ax1 = plt.subplots(figsize=(14, 6))

    # Plot Gold Price (แกน Y หลัก: ซ้าย)
    color = 'tab:orange'
    ax1.set_xlabel('Date')
    ax1.set_ylabel(f'{gold_col} (THB)', color=color)
    ax1.plot(merged_df.index, merged_df[gold_col], color=color, linewidth=2, label='Gold Price (THB)')
    ax1.tick_params(axis='y', labelcolor=color)
    ax1.grid(True, linestyle='--', alpha=0.6)

    # Plot CPI Index (แกน Y รอง: ขวา)
    ax2 = ax1.twinx()  # สร้างแกน Y ที่สองที่แชร์แกน X เดียวกัน
    color = 'tab:blue'
    ax2.set_ylabel(f'{cpi_trend_col} (Index)', color=color)
    ax2.plot(merged_df.index, merged_df[cpi_trend_col], color=color, linestyle='--', label=cpi_trend_col)
    ax2.tick_params(axis='y', labelcolor=color)

    plt.title(f'Time Series Comparison: {gold_col} vs. {cpi_trend_col}', fontsize=14)
    fig.tight_layout()
    plt.savefig('eda_gold_vs_cpi_timeseries_dualaxis.png')
    plt.show()
else:
    # แม้จะแก้ไขแล้ว ก็ยังคงการตรวจสอบเผื่อไว้
    print(f"ไม่พบหนึ่งในคอลัมน์ที่จำเป็นสำหรับการ Plot: {gold_col} หรือ {cpi_trend_col} (โปรดตรวจสอบการ Pivot ใน Cell 10)")
```

## 12. Scatter Plot (CPI vs. Gold_Next)
โค้ดนี้จะพล็อตความสัมพันธ์ระหว่าง CPI หมวดหมู่ย่อยที่สำคัญกับตัวแปรเป้าหมาย (gold_next) พร้อมแสดงเส้น Regression เพื่อให้เห็นแนวโน้มความสัมพันธ์เชิงเส้น

```python
# เลือกคอลัมน์ CPI หมวดหมู่ย่อยสำหรับการวิเคราะห์ความสัมพันธ์
cpi_scatter_col = 'หมวดอาหารและเครื่องดื่มไม่มีแอลกอฮอล์'
target_col = 'gold_next'

if cpi_scatter_col in merged_df.columns and target_col in merged_df.columns:
    
    plt.figure(figsize=(8, 6))
    
    # ใช้ regplot เพื่อแสดง Scatter Plot และเส้น Regression
    sns.regplot(
        x=merged_df[cpi_scatter_col],
        y=merged_df[target_col],
        scatter_kws={'alpha': 0.6},
        line_kws={'color': 'red'}
    )
    
    # ดึงค่า Correlation ที่คำนวณไว้ก่อนหน้า (หรือคำนวณซ้ำเพื่อยืนยัน)
    corr = merged_df[cpi_scatter_col].corr(merged_df[target_col])
    
    plt.title(f'Scatter Plot: {cpi_scatter_col} vs. {target_col}\n(Correlation: {corr:.4f})', fontsize=12)
    plt.xlabel(f'{cpi_scatter_col} (Index)', fontsize=10)
    plt.ylabel(f'{target_col} (THB)', fontsize=10)
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.tight_layout()
    plt.savefig('eda_cpi_vs_gold_next_scatter.png')
    plt.show()
else:
    print(f"ไม่พบหนึ่งในคอลัมน์ที่จำเป็นสำหรับการ Plot: {target_col} หรือ {cpi_scatter_col} (โปรดตรวจสอบการ Pivot ใน Cell 10)")
```
