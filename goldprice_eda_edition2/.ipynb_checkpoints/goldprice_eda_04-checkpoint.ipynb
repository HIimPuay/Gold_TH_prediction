## 1. Setup และการโหลดข้อมูลหลัก
โค้ดนี้จะนำเข้าไลบรารีที่จำเป็นทั้งหมด และโหลดไฟล์ feature_store.csv รวมถึงแปลงคอลัมน์ date ให้เป็น Time-Series Index

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# ตั้งค่าการแสดงผล
pd.set_option('display.max_columns', None)
plt.style.use('ggplot')
sns.set_style('whitegrid')

# การตั้งค่าฟอนต์ Tahoma เพื่อรองรับภาษาไทยในกราฟ
plt.rc('font', family='Tahoma')
plt.rcParams['axes.unicode_minus'] = False

# โหลด Feature Store (ไฟล์หลัก)
df = pd.read_csv('feature_store.csv')

# แปลง 'date' เป็น datetime และตั้งเป็น Index (จำเป็นสำหรับ Time Series Analysis)
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

print("--- Data Loading Complete ---")
```

## 2. โครงสร้างข้อมูลและการตรวจสอบคุณภาพเบื้องต้น
ตรวจสอบชนิดข้อมูล จำนวนแถว/คอลัมน์ และ Missing Values ก่อนทำการแก้ไข

```python
print("--- ข้อมูลเบื้องต้น (Structure) ---")
df.info()

print("\n--- ตัวอย่างข้อมูล 5 แถวแรก (Head) ---")
print(df.head())

print("\n--- การตรวจสอบ Missing Values (NaN) ---")
# ตรวจสอบ Missing Values ที่มักจะเกิดขึ้นในแถวเริ่มต้นของ Lagged/Rolling Features (Missing By Design)
nan_check = df.isnull().sum()[df.isnull().sum() > 0]
print(nan_check)

# --- จัดการ Missing by Design ---
# ทำการลบแถวที่มี NaN ทิ้ง (เกิดจาก Lag/Rolling Features ใน 3 แถวแรก) เพื่อให้ข้อมูลพร้อมสำหรับโมเดลอนุกรมเวลา
if not nan_check.empty:
    print("\n--- ทำการลบแถวที่มี NaN (แถวเริ่มต้น) เพื่อให้ข้อมูลพร้อมสำหรับโมเดลอนุกรมเวลา ---")
    df.dropna(inplace=True)
    print(f"จำนวนแถวที่เหลือหลังจาก dropna: {len(df)}")
```

## 3. สถิติเชิงพรรณนา (Descriptive Statistics) และการวิเคราะห์การกระจายตัว
แสดงค่าสถิติสำหรับตัวแปรหลักทั้งหมด โดยการ Transpose เพื่อความสะดวกในการอ่าน  
การเพิ่ม Skewness และ Kurtosis เป็นการวิเคราะห์ Non-Normality ที่สำคัญสำหรับข้อมูลทางการเงิน (เช่น _pct_change มักจะมี Kurtosis สูง, หรือ Fat Tails) ซึ่งจำเป็นต่อการทำนายความเสี่ยง

```python
print("--- สถิติเชิงพรรณนา (Descriptive Statistics) และการวิเคราะห์การกระจายตัว ---")
# เลือกคอลัมน์หลักที่มีความสำคัญ
key_cols = ['gold', 'gold_next', 'fx', 'cpi', 'oil', 'set', 'btc'] + [col for col in df.columns if '_pct_change' in col]

# แสดงผลแบบ Transposed
desc_stats = df[key_cols].describe().T

# --- เพิ่ม Skewness (ความเบ้) และ Kurtosis (ความโด่ง) ---
# Skewness บ่งบอกความสมมาตร, Kurtosis บ่งบอกความโด่ง (Fat Tails, ค่า > 3)
skewness = df[key_cols].skew()
kurtosis = df[key_cols].kurt()

# รวม Skewness และ Kurtosis เข้าไปในตารางสถิติ
desc_stats['skewness'] = skewness
desc_stats['kurtosis'] = kurtosis

print(desc_stats)
```

## 4. การกระจายตัวและ Outliers (Histograms & Box Plots)
แสดงภาพการกระจายตัวของตัวแปรหลักเพื่อทำความเข้าใจลักษณะของข้อมูล  
ใช้ Histogram เพื่อดูรูปร่างของ Distribution และ Box Plot เพื่อยืนยันการมีอยู่ของ Outliers

```python
visual_cols = ['gold', 'gold_next', 'fx', 'btc', 'gold_pct_change', 'btc_pct_change']

# พล็อต Histograms (KDE)
plt.figure(figsize=(15, 10))
for i, col in enumerate(visual_cols):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}', fontsize=10)
    plt.xlabel(col, fontsize=8)
    plt.ylabel('Frequency', fontsize=8)
plt.tight_layout()
plt.savefig('eda_distribution_plots.png')
plt.show()

# พล็อต Box Plots สำหรับตรวจสอบ Outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(visual_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=df[col].dropna())
    plt.title(f'Box Plot of {col}', fontsize=10)
    plt.ylabel(col, fontsize=8)
plt.tight_layout()
plt.savefig('eda_boxplot_plots.png')
plt.show()
```

## 5. Correlation Matrix (Heatmap)
วิเคราะห์ความสัมพันธ์เชิงเส้นตรง (Linear Correlation) เน้น Correlation บนตัวแปร % Change (Stationary Data) เพื่อหลีกเลี่ยง Spurious Correlation  
การใช้ % Change (Stationary Data) ในการวิเคราะห์ Correlation เป็นการหาความสัมพันธ์เชิงเส้นตรงที่แท้จริงในตลาดทุน เพื่อหลีกเลี่ยง Spurious Correlation ที่มักเกิดขึ้นในข้อมูลระดับราคา (Non-Stationary Data)

```python
# สร้าง Target Pct Change (ถ้ายังไม่มี)
df['gold_pct_change_next'] = df['gold_next'].pct_change() * 100

# เลือกคอลัมน์อัตราการเปลี่ยนแปลงทั้งหมด
pct_cols = [col for col in df.columns if '_pct_change' in col and col != 'gold_pct_change']
corr_cols = ['gold_pct_change_next'] + ['gold_pct_change'] + pct_cols

df_pct_corr = df[corr_cols].dropna()
corr_matrix_pct = df_pct_corr.corr()

# พล็อต Heatmap
plt.figure(figsize=(10, 8))
# ใช้ annot=True เพื่อแสดงค่า Correlation บน Heatmap ทันที
sns.heatmap(corr_matrix_pct, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, cbar_kws={'shrink': 0.8}, linecolor='black')
plt.title('Correlation Heatmap (Percentage Change Features) [Non-Spurious]', fontsize=14)
plt.tight_layout()
plt.savefig('eda_correlation_heatmap_pct.png')
plt.show()

print("\n--- Correlation กับ Target (gold_pct_change_next) [Stationary Correlation] ---")
# แสดงเฉพาะความสัมพันธ์กับตัวแปรเป้าหมายใหม่
target_corr = corr_matrix_pct['gold_pct_change_next']
if isinstance(target_corr, pd.DataFrame):
    target_corr = target_corr.iloc[:, 0]
print(target_corr.sort_values(ascending=False))
```

## 6. Time-Series Plotting (Trend and Volatility)
แสดงภาพอนุกรมเวลาเพื่อดูแนวโน้มและการเปลี่ยนแปลงของความผันผวน  
การทำ Plot Time Series เป็นขั้นตอนแรกในการระบุ Trend และ Volatility ด้วยสายตา  
การใช้ plt.xticks(rotation=45) ช่วยให้ป้ายชื่อวันที่บนแกน X ไม่ทับซ้อนกันและอ่านง่ายขึ้น

```python
ts_cols = ['gold', 'fx', 'btc', 'set']

plt.figure(figsize=(15, 12))
for i, col in enumerate(ts_cols):
    plt.subplot(4, 1, i + 1)
    df[col].plot(title=f'Time Series Plot of {col}', linewidth=1.5)
    plt.xlabel('Date')
    plt.ylabel(col)
    # Rotate x-labels for readability
    plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('eda_time_series_plots_improved.png')
plt.show()
```

## 7. Autocorrelation Function (ACF) และ PACF Plot
ตรวจสอบการพึ่งพาอาศัยกันตามเวลา (Autocorrelation/Partial Autocorrelation) เพื่อช่วยในการหาพารามิเตอร์ p และ q ของโมเดล ARIMA  
สำหรับการสร้างโมเดล ARIMA (p, d, q) PACF Plot (หา p) และ ACF Plot (หา q) เป็นเครื่องมือหลักในการเลือกพารามิเตอร์ที่เหมาะสม  
การพล็อตทั้งสองกราฟคู่กันสำหรับตัวแปรที่ Stationary แล้วเป็นขั้นตอนมาตรฐานของ Time Series Analysis

```python
print("--- PACF Plot สำหรับการหา AR Order (p) ---")

# PACF Plot สำหรับ Gold % Change (Stationary Data)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
# plot_pacf ใช้ในการหา AR Order (p)
plot_pacf(df['gold_pct_change'].dropna(), lags=30, ax=plt.gca(), title='PACF for gold_pct_change (AR Order p)')
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')

# ACF Plot สำหรับ Gold % Change (Stationary Data)
plt.subplot(1, 2, 2)
# plot_acf ใช้ในการหา MA Order (q)
plot_acf(df['gold_pct_change'].dropna(), lags=30, ax=plt.gca(), title='ACF for gold_pct_change (MA Order q)')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')

plt.tight_layout()
plt.savefig('eda_pacf_plots_improved.png')
plt.show()
```

## 8. Stationarity Check (ADF Test)
ตรวจสอบคุณสมบัติ Stationary ซึ่งจำเป็นสำหรับโมเดล ARIMA-type

ADF Test เป็นการทดสอบทางสถิติที่จำเป็นเพื่อยืนยันคุณสมบัติ Stationary  
ผลลัพธ์ยืนยันว่าข้อมูลระดับราคา (gold, fx) เป็น Non-Stationary (ต้องทำ Differencing) แต่ % Change เป็น Stationary แล้ว

```python
def adf_test_results(series):
    result = adfuller(series.dropna(), autolag='AIC')
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'P-value: {result[1]:.4f}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value:.4f}')
    if result[1] <= 0.05:
        print("=> สรุป: ข้อมูลมีคุณสมบัติ Stationary (Reject Null Hypothesis)")
    else:
        print("=> สรุป: ข้อมูลไม่มีคุณสมบัติ Stationary (Fail to Reject Null Hypothesis)")

print("--- ADF Test Results (Stationarity Check) ---")

print("\n[Gold Price]")
adf_test_results(df['gold'])

print("\n[FX Rate]")
adf_test_results(df['fx'])

print("\n[Gold Percentage Change]")
adf_test_results(df['gold_pct_change'])
```

## 9. การโหลดข้อมูล CPI ละเอียด (Feature Enrichment Prep)
CPI เป็นตัวแปรทางเศรษฐศาสตร์มหภาคที่เชื่อว่ามีความสัมพันธ์กับราคาทองคำ จำเป็นต้องโหลดและตรวจสอบโครงสร้างข้อมูล CPI รายละเอียดก่อนนำไปรวมกับข้อมูลหลัก

```python
# โหลดและตรวจสอบไฟล์ CPI_clean_for_supabase.csv
cpi_df = pd.read_csv('CPI_clean_for_supabase.csv')

# แปลง 'date' เป็น datetime
cpi_df['date'] = pd.to_datetime(cpi_df['date'])

print("--- ข้อมูล CPI ละเอียด (CPI_clean_for_supabase.csv) ---")
cpi_df.info()

print("\n--- ตัวอย่างข้อมูล 5 แถวแรก ---")
print(cpi_df.head())
```

## 10. การวิเคราะห์ CPI เชิงลึกและการหา Correlation (พร้อม Lagged Feature)
การสร้าง Lagged CPI Feature เป็นการทดสอบอำนาจในการทำนายในอนาคต (Predictive Power) และ CPI เป็นข้อมูลที่มีการอัปเดตช้า (Monthly) การใช้ค่าที่ Lag 30 วันจะจำลองสถานการณ์จริงที่ต้องใช้ข้อมูล CPI เดือนที่แล้วในการทำนายราคา Gold วันนี้

```python
# --- สร้าง Lagged CPI Feature และคำนวณ Correlation เพื่อหา Predictive Power ---

selected_cpi_codes = ['10000', '51000'] # เลือกเฉพาะหมวดหมู่ย่อยที่ไม่ใช่ 'รวมทุกรายการ'
cpi_subset = cpi_df[cpi_df['code'].astype(str).str.contains('|'.join(selected_cpi_codes))]

# --- 2. Pivot ตารางเพื่อให้หมวดหมู่เป็นคอลัมน์ ---
cpi_pivot = cpi_subset.pivot(index='date', columns='name_th', values='cpi_index')
cpi_pivot.columns.name = None

# --- 3. การแก้ไขปัญหา Mismatch Frequency: Resample และ Forward Fill (FFILL) ---
# ใช้วิธี asfreq('D') เพื่อสร้างดัชนีรายวัน และ ffill() เพื่อเติมค่าล่าสุด (CPI อัปเดตแบบ Step-wise)
cpi_daily = cpi_pivot.asfreq('D').ffill()

# --- 4. สร้าง Lagged CPI Feature ก่อน Join ---
# เลือก CPI หมวดอาหารฯ ที่มี Correlation สูง 
cpi_trend_col = 'หมวดอาหารและเครื่องดื่มไม่มีแอลกอฮอล์'
lagged_col_name = f'{cpi_trend_col}_Lag30D'
cpi_daily[lagged_col_name] = cpi_daily[cpi_trend_col].shift(30) # Lag 30 วัน (ประมาณ 1 เดือน)

print("--- ตัวอย่างข้อมูล CPI รายวันพร้อม Lagged Feature (5 แถวแรก) ---")
print(cpi_daily.head())

# --- 5. Inner Join กับ Feature Store ---
# Join CPI Daily (พร้อม Lag Feature) กับ DataFrame หลัก
merged_df = df.join(cpi_daily, how='inner')
merged_df.dropna(subset=['gold_next', lagged_col_name], inplace=True) # ทำความสะอาด NaN ที่เกิดจากการ Shift

# --- 6. คำนวณ Correlation ---
print("\n--- Correlation ของ CPI หมวดหมู่ย่อยกับ gold_next (หลัง FFILL) [พร้อม Lagged Feature] ---")

cpi_cols_to_check = [cpi_trend_col, lagged_col_name, 'ค่าโดยสารสาธารณะ']

for col in cpi_cols_to_check:
    if col in merged_df.columns:
         correlation_value = merged_df[col].corr(merged_df['gold_next'])
         print(f"CPI ({col}) Correlation: {correlation_value:.4f}")
```

## 11. Time-Series Comparison Plot (Dual-Axis Plot)
โค้ดนี้จะเปรียบเทียบแนวโน้มของราคาทองคำ (gold) กับดัชนีราคาผู้บริโภค (หมวดอาหารฯ) โดยใช้แกน Y สองแกน (Dual-Axis)  
Dual-Axis ช่วยให้เปรียบเทียบรูปแบบการเคลื่อนไหวร่วมกัน (Co-movement Pattern) ของตัวแปรที่มี Scale ต่างกันมากได้อย่างชัดเจน

```python
cpi_trend_col = 'หมวดอาหารและเครื่องดื่มไม่มีแอลกอฮอล์' 
gold_col = 'gold'

if cpi_trend_col in merged_df.columns and gold_col in merged_df.columns:
    
    fig, ax1 = plt.subplots(figsize=(14, 6))

    # Plot Gold Price (แกน Y หลัก: ซ้าย)
    color = 'tab:orange'
    ax1.set_xlabel('Date')
    ax1.set_ylabel(f'{gold_col} (THB)', color=color)
    ax1.plot(merged_df.index, merged_df[gold_col], color=color, linewidth=2, label='Gold Price (THB)')
    ax1.tick_params(axis='y', labelcolor=color)
    ax1.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(rotation=45) # Rotate x-labels for readability

    # Plot CPI Index (แกน Y รอง: ขวา)
    ax2 = ax1.twinx()  # สร้างแกน Y ที่สองที่แชร์แกน X เดียวกัน
    color = 'tab:blue'
    ax2.set_ylabel(f'{cpi_trend_col} (Index)', color=color)
    ax2.plot(merged_df.index, merged_df[cpi_trend_col], color=color, linestyle='--', label=cpi_trend_col)
    ax2.tick_params(axis='y', labelcolor=color)

    plt.title(f'Time Series Comparison: {gold_col} vs. {cpi_trend_col}', fontsize=14)
    fig.tight_layout()
    plt.savefig('eda_gold_vs_cpi_timeseries_dualaxis.png')
    plt.show()
```

## 12. Scatter Plot (CPI vs. Gold_Next)
โค้ดนี้จะพล็อตความสัมพันธ์ระหว่าง CPI หมวดหมู่ย่อยที่สำคัญกับตัวแปรเป้าหมาย (gold_next) พร้อมแสดงเส้น Regression  
Scatter Plot และเส้น Regression ใช้เพื่อยืนยันความสัมพันธ์เชิงเส้นตรง (Linearity) ระหว่างตัวแปรได้อย่างชัดเจนในเชิงกราฟ

```python
cpi_scatter_col = 'หมวดอาหารและเครื่องดื่มไม่มีแอลกอฮอล์'
target_col = 'gold_next'

if cpi_scatter_col in merged_df.columns and target_col in merged_df.columns:
    
    plt.figure(figsize=(8, 6))
    
    # ใช้ regplot เพื่อแสดง Scatter Plot และเส้น Regression
    sns.regplot(
        x=merged_df[cpi_scatter_col],
        y=merged_df[target_col],
        scatter_kws={'alpha': 0.6},
        line_kws={'color': 'red'}
    )
    
    # ดึงค่า Correlation ที่คำนวณไว้ก่อนหน้า (หรือคำนวณซ้ำเพื่อยืนยัน)
    corr = merged_df[cpi_scatter_col].corr(merged_df[target_col])
    
    plt.title(f'Scatter Plot: {cpi_scatter_col} vs. {target_col}\\n(Correlation: {corr:.4f})', fontsize=12)
    plt.xlabel(f'{cpi_scatter_col} (Index)', fontsize=10)
    plt.ylabel(f'{target_col} (THB)', fontsize=10)
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.tight_layout()
    plt.savefig('eda_cpi_vs_gold_next_scatter.png')
    plt.show()
```
