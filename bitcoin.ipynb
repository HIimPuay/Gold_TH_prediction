```python
import os
import pandas as pd
import yfinance as yf
from datetime import datetime

# ---------- CONFIG ----------
SAVE_PATH = "data/raw/bitcoin_history.csv"
SYMBOL = "BTC-USD"
START_DATE = "2023-01-01"
END_DATE = datetime.now().strftime("%Y-%m-%d")
# ----------------------------

def fetch_bitcoin_history():
    print(f"üìä ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏Ñ‡∏≤ Bitcoin ‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏ï‡πà {START_DATE} ‡∏ñ‡∏∂‡∏á {END_DATE} ...")
    df = yf.download(SYMBOL, start=START_DATE, end=END_DATE, progress=False)

    if df.empty:
        print("‚ùå ‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å Yahoo Finance")
        return None

    # ‚úÖ ‡∏£‡∏µ‡πÄ‡∏ã‡πá‡∏ï index ‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡πâ‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà
    df.reset_index(inplace=True)
    df.rename(columns={
        "Date": "date",
        "Open": "open",
        "High": "high",
        "Low": "low",
        "Close": "close",
        "Adj Close": "adj_close",
        "Volume": "volume"
    }, inplace=True)

    # ‚úÖ ‡πÄ‡∏û‡∏¥‡πà‡∏° timestamp ‡∏ï‡∏≠‡∏ô‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å
    df["timestamp"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # ‚úÖ ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏û‡∏¥‡πà‡∏° symbol ‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå CSV
    return df

def save_to_csv(df):
    os.makedirs(os.path.dirname(SAVE_PATH), exist_ok=True)
    df.to_csv(SAVE_PATH, index=False, encoding="utf-8-sig")
    print(f"‚úÖ ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• {len(df)} ‡πÅ‡∏ñ‡∏ß ‡∏•‡∏á‡πÑ‡∏ü‡∏•‡πå: {SAVE_PATH}")

def main():
    df = fetch_bitcoin_history()
    if df is not None and not df.empty:
        save_to_csv(df)

if __name__ == "__main__":
    main()
```

```python
import yfinance as yf
import pandas as pd

# ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Bitcoin ‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏ï‡πà‡∏õ‡∏µ 2023 - 2025
data = yf.download("BTC-USD", start="2023-01-01", end="2025-12-31", group_by="ticker")

# ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ multi-level columns ‡πÉ‡∏´‡πâ flatten ‡∏≠‡∏≠‡∏Å
if isinstance(data.columns, pd.MultiIndex):
    data.columns = [col[1] if col[1] else col[0] for col in data.columns]

# ‡∏£‡∏µ‡πÄ‡∏ã‡πá‡∏ï index ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ Date ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå
data.reset_index(inplace=True)

# ‡∏•‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Symbol ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ
if "Symbol" in data.columns:
    data.drop(columns=["Symbol"], inplace=True)

# ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏•‡∏á‡πÑ‡∏ü‡∏•‡πå CSV
output_path = "data/raw/bitcoin_history.csv"
data.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"‚úÖ ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢: {output_path}")
```
